### 1. Langauge Modelling
- Language Modelling is the task of predicting the next word in a sentence given the previous words.
- This is self supervised learning task. The model is trained on a large corpus of text data.

### 2. Supervised Fine Tuning (SFT)
- Supervised Fine Tuning is the task of fine tuning a pre-trained model on a custom dataset.
- The model is trained on a custom dataset with labels.



### 3. Preference Fine Tuning
- Preference Fine Tuning is the task of fine tuning a pre-trained model on a custom dataset with preference labels.
- The model is trained on a custom dataset with preference labels.

## Parameter-Efficient Fine-Tuning (PEFT)
- PEFT is a technique to fine-tune a pre-trained model with a small number of parameters.
- It has major components:
  - Adapter
  - LORA
  - QLORA

### Adapter
- Adapter is a small neural network that is added to the pre-trained model. It has a small number of parameters.
- Overall model size is increased by a small amount but fine tuning is done with a small number of parameters.

## LORA
- **LORA** is a technique to fine-tune a pre-trained model with a small number of parameters. 
- **Low-Rank Decomposition:** LoRA decomposes the weight matrices in neural network layers into two lower-rank matrices. This decomposition reduces the number of parameters that need to be fine-tuned, leading to efficient adaptation without compromising model performance.
- **Injecting Trainable Parameters:** The original weights are kept frozen, and only the low-rank matrices are trainable. These trainable matrices are injected into the neural network layers, capturing task-specific information during the fine-tuning process.
- **Combining Outputs:** During forward propagation, the output of the original weight matrix is combined with the output of the low-rank matrices. This combination allows the model to retain its pre-trained knowledge while adapting to new tasks with minimal parameter updates.

## QLORA
- **QLORA** is a technique to fine-tune a pre-trained model with a small number of parameters. But this time weights are quantized.
- **Hybrid Technique:** QLoRA combines quantization with low-rank adaptation to enhance parameter efficiency, reducing memory and computational requirements.
- **Resource-Constrained Fine-Tuning:** It enables efficient fine-tuning of large language models on limited hardware, making it feasible for real-world applications.
- **High Performance:** Despite the reduction in resources, QLoRA maintains high performance on downstream tasks by optimizing the adaptation process.


### Normalized Quantization
- **Normalized Quantization:** QLoRA uses normalized quantization to quantize the low-rank matrices. This technique ensures that the quantized weights are normalized to the same scale as the original weights, preserving the model's performance during fine-tuning.
